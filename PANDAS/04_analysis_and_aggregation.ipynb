# `Aggregate Functions` = Reduces a set of values into a single summary value.Used to summarize and analyze data.often used with groupby() function

># These are applied to whole Dataset

```python
print(df.mean(numeric_only=True)) # Find mean of only that values in Dataset that are numeric only
```

```python
print(df.min(numeric_only=True))
```

```python
print(df.max(numeric_only=True))
```

```python
print(df.count(numeric_only=True))
```

># Aggregate Functions on single column

```python
print(df["Age"].mean())
```

```python
print(df["Age"].median())
```

```python
print(df["Age"].min())
```

```python
print(df["Age"].mode())
```

```python
print(df["Age"].max())
```

```python
df.groupby("Pclass")["Fare"].mean()
```

```python
df.groupby("Pclass")["Fare"].min()
```

```python
df.groupby("Pclass")["Fare"].median()
```

```python
df.groupby("Pclass")["Fare"].max()
```

```python
df.groupby("Pclass")["Fare"].count()
```
