## Forward Fill — ffill
Meaning:

“Copy the last valid value above and use it to fill the missing one below.”

```python
dataa = {
    'Day': [1, 2, 3, 4, 5],
    'Temperature': [30, np.nan, np.nan, 33, np.nan]
}

df1 = pd.DataFrame(dataa)
```

```python
df1
```

```python
df1['Temperature'] = df1['Temperature'].ffill()
```

```python
df1
```

```python
df
```

## Backward Fill — bfill
Meaning:

“Copy the next valid value below and use it to fill the missing one above.”

```python
df1['Temperature'] = df1['Temperature'].bfill()
df1
```

```python
df1['Day'].replace(5 , 6)
```
