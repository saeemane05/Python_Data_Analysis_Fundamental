```python
import numpy as np
import time
import sys
```

```python
a  = [1 , 2, 3 ,4 ,5] * 1000000
start  = time.time()
for i in range(len(a)):
    a[i]  = a[i]*2
print("\nTime Taken for list:" , time.time()- start)
```

```python
a  = [1 , 2, 3 ,4 ,5] * 1000000
start  = time.time()
b = np.array(a)
B = b*2
print("\nTime taken for Numpy array:" ,time.time() - start)
```
