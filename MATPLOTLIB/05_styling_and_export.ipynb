```python
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
x = np.linspace(0, 10, 50)
y = np.sin(x)
```

```python
x
```

```python
y
```

```python
plt.style.use("seaborn-v0_8-whitegrid")
```

```python
plt.figure()
plt.plot(x, y, color="crimson", linestyle="--", linewidth=2, label="Styled Line")
plt.legend()
plt.title("Color, Line Style & Width")
plt.show()
```

```python
plt.figure()
plt.plot(x, y, marker="o", markersize=6, markerfacecolor="white")
plt.title("Marker Styling")
plt.show()
```

```python
plt.figure()
plt.plot(x, y)

plt.title("Large Bold Title", fontsize=14, fontweight="bold")
plt.xlabel("X Axis", fontsize=12)
plt.ylabel("Y Axis", fontsize=12)

plt.show()
```

```python
plt.figure()
plt.plot(x, y)

plt.grid(True, linestyle=":", linewidth=0.7, alpha=0.7)
plt.title("Custom Grid Style")

plt.show()
```

```python
plt.figure()
plt.plot(x, y)

plt.xlim(2, 8)
plt.ylim(-1.5, 1.5)

plt.xticks(fontsize=10, rotation=45)
plt.yticks(fontsize=10)

plt.title("Axis Limits & Tick Styling")
plt.show()
```

```python
plt.figure()
plt.plot(x, y, label="Sine")

plt.legend(
    loc="upper right",
    frameon=True,
    shadow=True,
    fontsize=10
)

plt.title("Styled Legend")
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(x, y)

ax.set_facecolor("#f5f5f5")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

ax.set_title("Clean Background & Spines")

plt.show()
```

```python
max_x = x[y.argmax()]
max_y = y.max()

plt.figure()
plt.plot(x, y)
plt.scatter(max_x, max_y)

plt.annotate(
    "Peak",
    xy=(max_x, max_y),
    xytext=(max_x + 1, max_y),
    arrowprops=dict(arrowstyle="->", color="black"),
    fontsize=10
)

plt.title("Styled Annotation")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
plt.plot(x, y)
plt.title("Layout Control Example")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(6, 4))
plt.plot(x, y)
plt.title("High Quality Export")

plt.savefig("styled_plot.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt

departments = ["HR", "IT", "Finance", "Medical"]
salaries = [30000, 50000, 40000, 45000]
```

```python
plt.figure()
plt.barh(departments, salaries)
plt.title("Horizontal Bar Style")
plt.xlabel("Salary")
plt.ylabel("Department")
plt.show()
```

```python
plt.figure()
plt.bar(departments, salaries, width=0.5, alpha=0.7)
plt.title("Bar Width & Transparency Style")
plt.show()
```

```python
plt.figure()
plt.bar(departments, salaries, color=["crimson", "teal", "gold", "purple"])
plt.title("Custom Colors Style")
plt.show()
```

```python
plt.figure()
plt.bar(departments, salaries, edgecolor="black", linewidth=1.2)
plt.title("Bar Edge Styling")
plt.show()
```

```python
plt.figure()
bars = plt.bar(departments, salaries)

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width() / 2,
        bar.get_height(),
        str(bar.get_height()),
        ha="center",
        va="bottom"
    )

plt.title("Bar Labels Style")
plt.show()
```
