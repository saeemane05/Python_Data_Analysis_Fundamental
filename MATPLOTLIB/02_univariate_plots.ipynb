```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
data = {
    "Salary": [250000 , 10000 , 35666, 450000 , 1000000 , 300000 , 600000]
}

df = pd.DataFrame(data)
```

```python
plt.style.use('grayscale')
plt.plot(df['Salary'] , color = 'red' , marker = 'o' , linestyle = '--' , linewidth = '1')
plt.grid()
plt.show()
```

# Line Plot (Trend Feel â€” Even for One Variable) .Good for showing order or time-based patterns.

```python
plt.figure(figsize = (6  , 6))
plt.style.use('grayscale')
plt.plot(df['Salary'] , marker = '*' ,markersize = '8' ,  linestyle  = '-' , linewidth = '1' , color = 'green')
plt.title("Salary Trend (Index vs Salary)")
plt.xlabel("Employee Index")
plt.ylabel("Salary")
plt.grid(True)
plt.show()
```

# Histogram (Distribution Story) - Shows how values are spread

```python
plt.figure(figsize = (6 , 6))
plt.style.use('grayscale')
plt.hist(df['Salary'] , bins = 10 , edgecolor = 'black')
plt.title("Salary Trend (Index vs Salary)")
plt.xlabel("Employee Index")
plt.ylabel("Salary")
plt.show()
```

```python
mean_salary = df['Salary'].mean()

plt.figure(figsize=(6, 3))
plt.hist(df['Salary'], bins=20)
plt.axvline(mean_salary, linestyle='--', label=f"Mean = {int(mean_salary)}")
plt.title("Salary Distribution with Mean")
plt.legend()
plt.show()
```

# Boxplot (Outlier Detector) - Great for interview talk.

##  Box Plot (Box-and-Whisker Plot)

A **box plot** shows how your data is spread and where most values lie.

### ðŸ”¹ Parts

- **Min** â†’ Smallest value in the data  
- **Q1 (Lower Quartile)** â†’ 25% of data is below this  
- **Median (Q2)** â†’ Middle value of the data  
- **Q3 (Upper Quartile)** â†’ 75% of data is below this  
- **Max** â†’ Largest value in the data  

### ðŸ”¹ Box & Whiskers

- **Box** â†’ From **Q1 to Q3**  
- **IQR (Interquartile Range)** â†’  

`IQR = Q3 - Q1`
- **Line inside box** â†’ Median  
- **Whiskers** â†’ Lines from box to min and max  

### ðŸ”¹ Why Use It?

- See **spread of data**
- Find **outliers**
- Compare **distributions quickly**

```python
plt.figure(figsize=(6, 6))
plt.boxplot(df["Salary"], vert=False)  # vert =False makes the plot Horizontal
plt.title("Salary Boxplot (Outliers Visible)")
plt.xlabel("Salary")
plt.show()
```

```python
plt.figure(figsize=(6, 6))
plt.boxplot(df["Salary"], vert=True)  
plt.title("Salary Boxplot (Outliers Visible)")
plt.xlabel("Salary")
plt.show()
```

`Adding an Outlier to the boxplot`

```python
df.loc[6] = [0]  # The circular dot is an outlier
plt.boxplot(df['Salary'])
plt.show()
```

`Removing this Outlier`

```python
df.drop(index = 6 , inplace = True)
```

# Pie Chart (Category Frequency)


```python
departments = ["HR", "IT", "Finance", "Marketing", "Sales"]
counts = [45, 70, 40, 25, 20]

plt.figure(figsize=(4, 4))
plt.pie(counts, labels=departments, autopct='%1.1f' , explode=(0 , 0.1 , 0 , 0 , 0 )) # autopct for displaying percentage , #explode for the partition effect of a section 
plt.title("Employee Distribution by Department")
plt.axis('equal') # makes pie chart in actual circle

plt.show()
```

# Count Plot (Bar-Style Frequency)

```python
plt.figure(figsize=(6, 6))
plt.bar(departments, counts , edgecolor  = 'black' , linestyle = '--' , color = ['green' , 'cyan' , 'blue' , 'orange' , 'plum'])
plt.title("Department Counts")
plt.xlabel("Department")
plt.ylabel("Number of Employees")
plt.show()
```

```python
plt.style.use('seaborn-v0_8-whitegrid')

plt.figure(figsize=(6, 3))
plt.hist(df['Salary'], bins=25)
plt.title("Styled Salary Distribution")
plt.xlabel("Salary")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(6, 1))
plt.plot(df['Salary'], np.zeros_like(df["Salary"]), '|')
plt.title("Salary Rug Plot (Density View)")
plt.yticks([])
plt.show()
```
