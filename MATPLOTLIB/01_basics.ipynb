# Matplotlib Basics

## What is Matplotlib?
Matplotlib is a **Python plotting library** used to create **static, animated, and interactive visualizations**.  
It is widely used in **data science, machine learning, research, and reporting** for making high-quality graphs and charts.

---

## Who Invented Matplotlib?
Matplotlib was created by **John D. Hunter** in **2002** while he was working on scientific data visualization for neuroscience research.

---

## Why Use Matplotlib?
- Works smoothly with **NumPy and Pandas**
- Produces **publication-quality plots**
- Highly **customizable**
- Supported by a large **open-source community**

---

## Core Components of a Plot
- **Figure** → The whole canvas (window/image)
- **Axes** → The actual plotting area
- **Axis** → The x and y scale and ticks
- **Title & Labels** → Text that explains the plot
- **Legend** → Explains colors/lines

> # Pyplot API

# First Plot — `plt.plot()`

```python
x = [1 ,2 , 3 , 4 , 5]
y = [10 , 20 , 30 , 40 , 50]
plt.plot(x , y , 'c') #(1 , 10) , (2 , 20) , (3 , 30)......
plt.grid()  
plt.show()
```

# Add Labels & Title

```python
plt.plot(x , y , 'g')
plt.title("Simple Line Plot")
plt.xlabel("X Axis")
plt.ylabel("Y Axis")
plt.grid()
plt.show()
```

># Object Oriented API

```python
fig , axes  = plt.subplots()
axes.plot(x , y , 'r')
axes.set_title("OO API Line Plot")
axes.set_xlabel("X Axis")
axes.set_ylabel("Y Axis")
plt.grid()
plt.show()
```

# Multiple Lines (Same Plot)

```python
y2 = [2 , 4 , 6 , 8 , 10]
plt.plot(x , y , "r" , label = "Line 1")
plt.plot(x , y2 , "g" , label = "Line 2")
plt.grid()
plt.legend()
plt.show()
```

# Save the Figure

```python
y2 = [2 , 4 , 6 , 8 , 10]
plt.plot(x , y , "r" , label = "Line 1")
plt.plot(x , y2 , "g" , label = "Line 2")
plt.title("Saved Plot")
plt.grid()
plt.legend()
plt.show()
```
