```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.linspace(0, 10, 100)

fig, axes = plt.subplots(2, 2, figsize=(8, 6))

axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sine")

axes[0, 1].plot(x, np.cos(x))
axes[0, 1].set_title("Cosine")

axes[1, 0].plot(x, np.tan(x))
axes[1, 0].set_title("Tangent")

axes[1, 1].plot(x, np.exp(x / 5))
axes[1, 1].set_title("Exponential")

plt.tight_layout()
plt.show()
```

```python
fig, axes = plt.subplots(2, 1, figsize=(6, 6), sharex=True)

axes[0].plot(x, np.sin(x))
axes[0].set_title("Sine")

axes[1].plot(x, np.cos(x))
axes[1].set_title("Cosine")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize = (7 , 6))

plt.subplot(3, 1, 1)
plt.plot(df["Salary"], marker='o')
plt.title("Salary Trend")
plt.ylabel("Salary")
```

```python
plt.figure(figsize = (7 , 6))

plt.subplot(3, 1, 2)
plt.plot(df["Age"], marker='o')
plt.title("Age Trend")
plt.ylabel("Age")
```

```python
dept_mean = df.groupby("Department")["Salary"].mean()

plt.subplot(3, 1, 3)
plt.bar(dept_mean.index, dept_mean.values)
plt.title("Average Salary by Department")
plt.xlabel("Department")
plt.ylabel("Avg Salary")

plt.tight_layout()
plt.show()
```

```python

from sklearn.datasets import load_iris
import pandas as pd

# Load data
iris = load_iris()
df = pd.DataFrame(iris.data, columns=iris.feature_names)

# Create 2x2 grid
fig, axes = plt.subplots(2, 2, figsize=(7, 6))

# 1. Sepal Length vs Sepal Width
axes[0, 0].scatter(df["sepal length (cm)"], df["sepal width (cm)"])
axes[0, 0].set_title("Sepal L vs W")
axes[0, 0].set_xlabel("Sepal Length")
axes[0, 0].set_ylabel("Sepal Width")

# 2. Petal Length vs Petal Width
axes[0, 1].scatter(df["petal length (cm)"], df["petal width (cm)"])
axes[0, 1].set_title("Petal L vs W")
axes[0, 1].set_xlabel("Petal Length")
axes[0, 1].set_ylabel("Petal Width")

# 3. Sepal Length Boxplot
axes[1, 0].boxplot(df["sepal length (cm)"])
axes[1, 0].set_title("Sepal Length Box")
axes[1, 0].set_ylabel("Sepal Length")

# 4. Petal Length Boxplot
axes[1, 1].boxplot(df["petal length (cm)"])
axes[1, 1].set_title("Petal Length Box")
axes[1, 1].set_ylabel("Petal Length")

plt.tight_layout()
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(
    df["sepal length (cm)"],
    df["sepal width (cm)"],
    df["petal length (cm)"]
)

ax.set_title("3D Plot: Sepal vs Petal (Iris)")
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Sepal Width (cm)")
ax.set_zlabel("Petal Length (cm)")

plt.show()
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

ax.scatter(
    df["sepal length (cm)"],
    df["sepal width (cm)"],
    df["petal length (cm)"],
    c = df["petal width (cm)"]
)

ax.set_title("3D Plot (Colored by Petal Width)")
ax.set_xlabel("Sepal Length (cm)")
ax.set_ylabel("Sepal Width (cm)")
ax.set_zlabel("Petal Length (cm)")

plt.show()
```
