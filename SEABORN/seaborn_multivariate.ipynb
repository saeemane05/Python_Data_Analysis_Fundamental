# Heatmap


A heatmap is a visualization that represents data values as colors in a matrix, where color intensity shows the magnitude of each value, making patterns, correlations, and extremes easy to spot at a glance

```python
columns = ['bill_length_mm' , "bill_depth_mm" , 'flipper_length_mm' , 'body_mass_g']
```

```python
sns.heatmap(data = penguins[columns].corr() , annot = True)
```

```python
sns.heatmap(data = penguins[columns].corr() , cmap = 'Blues' , vmin= -0.2 , annot = True)
```

```python
sns.heatmap(data = penguins[columns].corr() , cmap = 'Blues' , vmin= -0.2, linewidth = 2 , annot = True)
```

```python
sns.rugplot(data =  penguins , x = 'flipper_length_mm', y = 'body_mass_g' , hue = 'sex' , palette = 'Set2' , height = 0.2)
```

```python
sns.pairplot(data =  penguins , hue = 'sex' , palette = "Set2" , diag_kind = 'kde')
```

# `pairgrid` = subplot grid for plotiing pairwise relationships in a dataset|

```python
graph = sns.PairGrid(data = penguins , hue = 'sex' , palette = "Set2")
graph.map_upper(sns.scatterplot) # upper traingle
graph.map_lower(sns.kdeplot) # lower triangle
graph.map_diag(sns.histplot) # diagonally
```
