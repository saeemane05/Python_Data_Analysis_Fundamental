# Univariate Data Analysis with Seaborn

This notebook focuses on **univariate analysis**, which means analyzing **one variable at a time** to understand:

- Distribution
- Central tendency
- Spread
- Outliers
- Shape of data (skewness, peaks, tails)

We use **Seaborn + Matplotlib + Pandas** to visualize and interpret patterns in real-world data.

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib.inline
```

```python
sns.get_dataset_names()
```

```python
penguins = sns.load_dataset('penguins')
penguins.head(7)
```

```python
penguins['species'].value_counts()
```

```python
penguins['island'].value_counts()
```

```python
penguins.dtypes
```

```python
sns.stripplot(data = penguins , x = 'species' , y = 'island')
```

```python
sns.stripplot(data = penguins , x = 'species' , y = 'body_mass_g')
```

```python
sns.set_context = 'notebook'
sns.histplot(data = penguins , x = 'body_mass_g' , hue = 'sex' , multiple = 'stack')
```
