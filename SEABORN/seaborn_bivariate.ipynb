```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib.inline
```

```python
penguins = sns.load_dataset('penguins')
penguins.head(7)
```

```python
penguins.dtypes
```

```python
sns.scatterplot(data = penguins , x = 'flipper_length_mm' , y = 'body_mass_g')
```

```python
sns.regplot(data = penguins , x = 'body_mass_g' , y = 'flipper_length_mm')
```

```python
sns.set_style('whitegrid')
sns.jointplot(data = penguins , x = 'body_mass_g' , y = 'flipper_length_mm' , color = 'deeppink' , palette = ['pink' , 'skyblue'])
```

```python
sns.set_style('whitegrid')
sns.barplot(data = penguins , x = 'species' , y = 'flipper_length_mm' , color = 'deeppink' , palette = ['pink' , 'skyblue'])
```

```python
sns.countplot(data =  penguins , x = 'species', color = 'teal')
```

```python
sns.boxplot(data =  penguins , x = 'species', color = 'maroon')
```

```python
sns.boxplot(
    data=penguins,
    x="species",
    y="body_mass_g",
    palette=["orange", "blue", "green"]
)
```

```python
sns.violinplot(data =  penguins , x = 'species', hue = 'sex' , split = True)
```
